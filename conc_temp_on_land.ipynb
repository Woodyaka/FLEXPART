```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Percentage of time spent below the boundary layer and above land (over 14 days)
percent_below_bl = {
    'C320r2t': 22.41,  
    'C320r1t': 20.30,  
    'C319r1t': 2.65,   
    'C319r5t': 1.66,   
    'C320r3t': 23.55,  
    'C320r4t': 6.47,   
    'C321r3t': 7.61,   
    'C321r1t': 1.81,   
    'C324r5t': 8.44,   
    'C322r2t': 6.67,   
    'C324r1t': 2.25,   
    'C324r2t': 1.96,   
    'C322r1t': 5.85,   
    'C323r4t': 7.56,   
    'C325r2t': 5.74,   
    'C325r4t': 11.35,  
    'C325r1t': 5.71,   
    'C326r1t': 4.19,   
    'C327r1t': 14.65,  
    'C328r2t': 3.22,   
    'C330r2t': 3.10,   
    'C327r2t': 13.11,  
    'C329r2t': 9.23,   
    'C329r3t': 2.32,   
    'C330r1t': 1.78,   
    'C328r1t': 15.31,  
    'C329r1t': 9.32,  
    'C331r3t': 2.05,   
    'C327r4t': 13.15,  
    'C331r4t': 5.19,   
    'C326r2t': 1.57,   
    'C332r1t': 2.38,   
    'C333r1t': 2.67,   
    'C332r2t': 2.63,   
    'C333r2t': 2.81,   
    'C333r3t': 2.47,   
}
# Only for points west of 20 degrees west
# percent_below_bl = {
#     'C320r2t' :  38.09,
#     'C320r1t' :  35.75,
#     'C319r1t' :  10.12,
#     'C319r5t' :  3.36,
#     'C320r3t' :  33.20,
#     'C320r4t' :  9.23,
#     'C321r3t' :  21.99,
#     'C321r1t' :  5.28,
#     'C324r5t' :  17.18,
#     'C322r2t' :  16.86,
#     'C324r1t' :  9.09,
#     'C324r2t' :  7.38,
#     'C322r1t' :  16.79,
#     'C323r4t' :  16.16,
#     'C325r2t' :  16.56,
#     'C325r4t' :  14.51,
#     'C325r1t' :  16.51,
#     'C326r1t' :  12.90,
#     'C327r1t' :  24.34,
#     'C328r2t' :  4.98,
#     'C330r2t' :  9.02,
#     'C327r2t' :  22.36,
#     'C329r2t' :  24.99,
#     'C329r3t' :  0.37,
#     'C330r1t' :  5.44,
#     'C328r1t' :  35.38,
#     'C329r1t' :  19.90,
#     'C331r3t' :  6.38,
#     'C327r4t' :  25.84,
#     'C331r4t' :  9.90,
#     'C326r2t' :  4.56,
#     'C332r1t' :  7.20,
#     'C333r1t' :  7.48,
#     'C332r2t' :  7.63,
#     'C333r2t' :  6.89,
#     'C333r3t' :  5.41
# }
# Only for points west of 20 degrees west in the first 7 days
# percent_below_bl = {
#     'C320r2t':  27.68,
#     'C320r1t':  26.42,
#     'C319r1t':  9.53,
#     'C319r5t':  14.96,
#     'C320r3t':  28.62,
#     'C320r4t':  17.35,
#     'C321r3t':  14.67,
#     'C321r1t':  16.72,
#     'C324r5t':  13.67,
#     'C322r2t':  14.86,
#     'C324r1t':  6.34,
#     'C324r2t':  4.20,
#     'C322r1t':  12.50,
#     'C323r4t':  16.21,
#     'C325r2t':  4.13,
#     'C325r4t':  17.93,
#     'C325r1t':  3.87,
#     'C326r1t':  1.60,
#     'C327r1t':  12.00,
#     'C328r2t':  8.20,
#     'C330r2t':  7.27,
#     'C327r2t':  11.20,
#     'C329r2t':  18.60,
#     'C329r3t':  1.23,
#     'C330r1t':  6.90,
#     'C328r1t':  14.52,
#     'C329r1t':  14.47,
#     'C331r3t':  7.08,
#     'C327r4t':  20.56,
#     'C331r4t':  9.50,
#     'C326r2t':  3.53,
#     'C332r1t':  9.53,
#     'C333r1t':  7.04,
#     'C332r2t':  9.90,
#     'C333r2t':  6.60,
#     'C333r3t':  6.35

# }
# Percentage of time spent below the boundary layer and above land (over 14 days) (aerotracer)
# percent_below_bl = {
#     'C320r2t':  27.68,
#     'C320r1t':  26.42,
#     'C319r1t':  9.53,
#     'C319r5t':  14.96,
#     'C320r3t':  28.62,
#     'C320r4t':  17.35,
#     'C321r3t':  14.67,
#     'C321r1t':  16.72,
#     'C324r5t':  13.67,
#     'C322r2t':  14.86,
#     'C324r1t':  6.34,
#     'C324r2t':  4.20,
#     'C322r1t':  12.50,
#     'C323r4t':  16.21,
#     'C325r2t':  4.13,
#     'C325r4t':  17.93,
#     'C325r1t':  3.87,
#     'C326r1t':  1.60,
#     'C327r1t':  12.00,
#     'C328r2t':  8.20,
#     'C330r2t':  7.27,
#     'C327r2t':  11.20,
#     'C329r2t':  18.60,
#     'C329r3t':  1.23,
#     'C330r1t':  6.90,
#     'C328r1t':  14.52,
#     'C329r1t':  14.47,
#     'C331r3t':  7.08,
#     'C327r4t':  20.56,
#     'C331r4t':  9.50,
#     'C326r2t':  3.53,
#     'C332r1t':  9.53,
#     'C333r1t':  7.04,
#     'C332r2t':  9.90,
#     'C333r2t':  6.60,
#     'C333r3t':  6.35

# }



# Read the Excel file
df = pd.read_excel('mphase_ER_CAOs.xlsx')

# Get the temperature data from the first column
temperature = df.iloc[:, 0]

# Normalize for consistent color mapping
all_percentages = list(percent_below_bl.values())
norm = plt.Normalize(vmin=min(all_percentages), vmax=max(all_percentages))

# Function to create a combined plot
def create_combined_plot(data_dict, title):
    fig, ax = plt.subplots(figsize=(12, 8))
    cmap = plt.cm.viridis
    
    all_y_data = []
    
    for run, percent in data_dict.items():
        # Filter columns that match the run and end with '_NT'
        matching_columns = [col for col in df.columns if run in col and col.endswith('_NT')]
        if matching_columns:
            for col in matching_columns:
                all_y_data.extend(df[col])
                
                color = cmap(norm(percent))
                ax.plot(temperature, df[col], 'o-', c=color, label=run)
        else:
            print(f"No matching column found for run {run} ending with '_NT'")
    
    if not all_y_data:
        print(f"No data plotted for {title}. Check if run names match column names.")
        return
    
    ax.set_xlabel('Temperature')
    ax.set_ylabel(r'$N_{INP}/L^{-1}$')
    ax.set_yscale('log')
    
    # Set axis limits
    ax.set_xlim(-37.5, -5)
    y_min = max(1e-3, np.min(all_y_data))
    y_max = max(1e1, np.max(all_y_data))
    ax.set_ylim(y_min, y_max)
    
    ax.set_title(f'{title} Run Concentrations vs Temperature')
    ax.grid(True)
    
    # Add colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, ax=ax)
    cbar.set_label('Percentage of time spent over land')
    
    ax.legend(loc='center left')
    
    plt.tight_layout()
    plt.show()
    #plt.savefig(fname=f'time_spent_over_land_run_concentrations_vs_Temperature.png', dpi=300, format='png')
    #plt.close()


print('The errors are fine, these are non CAOs!')

# Create a combined plot
create_combined_plot(percent_below_bl, '')
```
