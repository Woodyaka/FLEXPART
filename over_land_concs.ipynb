```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Percentage of time spent below the boundary layer and above land (over 14 days)
percent_below_bl = {
    'C320r2t': 22.41,  
    'C320r1t': 20.30,  
    'C319r1t': 2.65,   
    'C319r5t': 1.66,   
    'C320r3t': 23.55,  
    'C320r4t': 6.47,   
    'C321r3t': 7.61,   
    'C321r1t': 1.81,   
    'C324r5t': 8.44,   
    'C322r2t': 6.67,   
    'C324r1t': 2.25,   
    'C324r2t': 1.96,   
    'C322r1t': 5.85,   
    'C323r4t': 7.56,   
    'C325r2t': 5.74,   
    'C325r4t': 11.35,  
    'C325r1t': 5.71,   
    'C326r1t': 4.19,   
    'C327r1t': 14.65,  
    'C328r2t': 3.22,   
    'C330r2t': 3.10,   
    'C327r2t': 13.11,  
    'C329r2t': 9.23,   
    'C329r3t': 2.32,   
    'C330r1t': 1.78,   
    'C328r1t': 15.31,  
    'C329r1t': 9.32,  
    'C331r3t': 2.05,   
    'C327r4t': 13.15,  
    'C331r4t': 5.19,   
    'C326r2t': 1.57,   
    'C332r1t': 2.38,   
    'C333r1t': 2.67,   
    'C332r2t': 2.63,   
    'C333r2t': 2.81,   
    'C333r3t': 2.47,   
}


# Only for points west of 20 degrees west
# percent_below_bl = {
#     'C320r2t' :  38.09,
#     'C320r1t' :  35.75,
#     'C319r1t' :  10.12,
#     'C319r5t' :  3.36,
#     'C320r3t' :  33.20,
#     'C320r4t' :  9.23,
#     'C321r3t' :  21.99,
#     'C321r1t' :  5.28,
#     'C324r5t' :  17.18,
#     'C322r2t' :  16.86,
#     'C324r1t' :  9.09,
#     'C324r2t' :  7.38,
#     'C322r1t' :  16.79,
#     'C323r4t' :  16.16,
#     'C325r2t' :  16.56,
#     'C325r4t' :  14.51,
#     'C325r1t' :  16.51,
#     'C326r1t' :  12.90,
#     'C327r1t' :  24.34,
#     'C328r2t' :  4.98,
#     'C330r2t' :  9.02,
#     'C327r2t' :  22.36,
#     'C329r2t' :  24.99,
#     'C329r3t' :  0.37,
#     'C330r1t' :  5.44,
#     'C328r1t' :  35.38,
#     'C329r1t' :  19.90,
#     'C331r3t' :  6.38,
#     'C327r4t' :  25.84,
#     'C331r4t' :  9.90,
#     'C326r2t' :  4.56,
#     'C332r1t' :  7.20,
#     'C333r1t' :  7.48,
#     'C332r2t' :  7.63,
#     'C333r2t' :  6.89,
#     'C333r3t' :  5.41
# }
# Only for points west of 20 degrees west in the first 7 days
# percent_below_bl = {
#     'C320r2t':  27.68,
#     'C320r1t':  26.42,
#     'C319r1t':  9.53,
#     'C319r5t':  14.96,
#     'C320r3t':  28.62,
#     'C320r4t':  17.35,
#     'C321r3t':  14.67,
#     'C321r1t':  16.72,
#     'C324r5t':  13.67,
#     'C322r2t':  14.86,
#     'C324r1t':  6.34,
#     'C324r2t':  4.20,
#     'C322r1t':  12.50,
#     'C323r4t':  16.21,
#     'C325r2t':  4.13,
#     'C325r4t':  17.93,
#     'C325r1t':  3.87,
#     'C326r1t':  1.60,
#     'C327r1t':  12.00,
#     'C328r2t':  8.20,
#     'C330r2t':  7.27,
#     'C327r2t':  11.20,
#     'C329r2t':  18.60,
#     'C329r3t':  1.23,
#     'C330r1t':  6.90,
#     'C328r1t':  14.52,
#     'C329r1t':  14.47,
#     'C331r3t':  7.08,
#     'C327r4t':  20.56,
#     'C331r4t':  9.50,
#     'C326r2t':  3.53,
#     'C332r1t':  9.53,
#     'C333r1t':  7.04,
#     'C332r2t':  9.90,
#     'C333r2t':  6.60,
#     'C333r3t':  6.35

# }
# Percentage of time spent below the boundary layer and above land (over 14 days) (aerotracer)
# percent_below_bl = {
#     'C320r2t':  27.68,
#     'C320r1t':  26.42,
#     'C319r1t':  9.53,
#     'C319r5t':  14.96,
#     'C320r3t':  28.62,
#     'C320r4t':  17.35,
#     'C321r3t':  14.67,
#     'C321r1t':  16.72,
#     'C324r5t':  13.67,
#     'C322r2t':  14.86,
#     'C324r1t':  6.34,
#     'C324r2t':  4.20,
#     'C322r1t':  12.50,
#     'C323r4t':  16.21,
#     'C325r2t':  4.13,
#     'C325r4t':  17.93,
#     'C325r1t':  3.87,
#     'C326r1t':  1.60,
#     'C327r1t':  12.00,
#     'C328r2t':  8.20,
#     'C330r2t':  7.27,
#     'C327r2t':  11.20,
#     'C329r2t':  18.60,
#     'C329r3t':  1.23,
#     'C330r1t':  6.90,
#     'C328r1t':  14.52,
#     'C329r1t':  14.47,
#     'C331r3t':  7.08,
#     'C327r4t':  20.56,
#     'C331r4t':  9.50,
#     'C326r2t':  3.53,
#     'C332r1t':  9.53,
#     'C333r1t':  7.04,
#     'C332r2t':  9.90,
#     'C333r2t':  6.60,
#     'C333r3t':  6.35
# }


df = pd.read_excel('mphase_ER_CAOs.xlsx')

# Get the temperature data from the first column
temperature = df.iloc[:, 0]

# Function to create a plot with data split by percentage criteria
def create_plot(data_dict, title):
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Separate the runs based on the percentage criterion 
    over_5 = {k: v for k, v in data_dict.items() if v > 7.15} # This percentage is of 14 days (7.15% ~ 1 day)
    under_5 = {k: v for k, v in data_dict.items() if v < 7.15}
    
    # List to keep track of color and label for the two groups
    groups = [(over_5, 'Over 24 hours over land', 'r'), (under_5, 'Less than 24 hours over land', 'b')]
    
    for group, label, color in groups:
        if not group:
            continue
        
        # Initialize arrays to store the median INP concentration and error bars
        median_inp_conc = np.zeros(len(temperature))
        lower_errors = np.zeros(len(temperature))
        upper_errors = np.zeros(len(temperature))
        
        # Dictionaries to store INP concentration values for each temperature
        inp_conc_dict = {temp: [] for temp in temperature}
        
        # Loop through each run in the 'group' dictionary
        for run in group.keys():
            # Find columns in the DataFrame that match the current run and end with '_NT'
            matching_columns = [col for col in df.columns if run in col and col.endswith('_NT')]
            
            # Loop through each matching column
            for col in matching_columns:
                # Get the INP concentration data from the DataFrame
                y_data = df[col].values
                
                # Store INP concentration values for each temperature
                for i, temp in enumerate(temperature):
                    inp_conc_dict[temp].append(y_data[i])
        
        # Calculate the median INP concentration and error bars for each temperature
        for i, temp in enumerate(temperature):
            conc_values = inp_conc_dict[temp]
            median_inp_conc[i] = np.median(conc_values)
            lower_quartile = np.percentile(conc_values, 25)
            upper_quartile = np.percentile(conc_values, 75)
            lower_errors[i] = median_inp_conc[i] - lower_quartile
            upper_errors[i] = upper_quartile - median_inp_conc[i]
        
        # Plot median INP concentration with asymmetric error bars
        ax.errorbar(temperature, median_inp_conc, yerr=[lower_errors, upper_errors], fmt='o-', color=color,
                    label=f'{label}', linewidth=2, capsize=4)
    
    # Set axis labels and scales
    ax.set_xlabel('Temperature')
    ax.set_ylabel(r'$N_{INP}/L^{-1}$')
    ax.set_yscale('log')
    
    # Set y-axis limits based on the plotted data
    y_min = np.min(median_inp_conc - lower_errors)
    y_max = np.max(median_inp_conc + upper_errors)
    
    ax.set_title(f'Time spent over land and below bl run concentrations vs Temperature')
    ax.grid(True)
    
    # Create legend
    ax.legend(loc='best')
    
    plt.tight_layout()
    plt.show()
    # plt.savefig(fname=f'time_spent_over_land_run_concentrations_vs_Temperature_7_days_west_20W_24_hours.png', dpi=300, format='png')
    # plt.close()

# Create plot 
create_plot(percent_below_bl, 'Boundary Layer')
```
